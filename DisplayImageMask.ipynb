
#Display of images and corresponding masks

```
from google.colab import drive
drive.mount('/content/drive/', force_remount=False)
from google.colab.patches import cv2_imshow
import os 
import cv2
i=0
dir_img="CameraRGB/"
dir_mask="CameraSeg/"

for dir in ['/content/drive/My Drive/Colab Notebooks/dataset/lyft-udacity-challenge/resized/dataB/']:
 for file in os.listdir(dir+ dir_img):
   i=i+1
   img=cv2.imread(dir+dir_img+file)
   cv2_imshow(img)
   mask=cv2.imread(dir+dir_mask+file)
   cv2_imshow( mask*50)
   if i==20:
      break
    
```
